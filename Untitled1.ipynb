### 1

```python


# 构建订单失败的原因分类
data_orders['failure_reason'] = data_orders.apply(lambda row: 'Cancelled by Customer Before Driver Assignment' if row['order_status_key'] == 4 and row['is_driver_assigned_key'] == 0 else 
                                            'Cancelled by Customer After Driver Assignment' if row['order_status_key'] == 4 and row['is_driver_assigned_key'] == 1 else 
                                            'Rejected by System', axis=1)

# 计算每个原因的订单数量
failure_reason_counts = data_orders['failure_reason'].value_counts()

# 绘制条形图
plt.figure(figsize=(12, 6))  # 设置图形的尺寸
failure_reason_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'orange'])  # 设置条形颜色

# 添加网格线
plt.grid(axis='y', linestyle='--', linewidth=0.5)

# 添加标题和标签
plt.title('Distribution of Orders by Failure Reason', fontsize=16)
plt.xlabel('Reason for Order Failure', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)

# 旋转X轴标签以便于阅读
plt.xticks(rotation=45, ha='right')  # ha='right'确保标签不会互相覆盖



plt.tight_layout()

# 显示图形
plt.show()
```

From the bar chart, we can deduce the following insights regarding the distribution of failed orders based on their reasons for failure:
1. The majority of orders are canceled before a driver is assigned, likely due to changes in customer plans or dissatisfaction with waiting times.
2. The number of orders rejected by the system is moderate, suggesting potential issues with driver matching or availability.
3. The least number of orders are canceled after a driver has been assigned, indicating that passengers are more inclined to wait once a driver is en route, which reflects higher efficiency and customer satisfaction at this stage of the system.

Based on these insights, Gett might consider measures to expedite the matching process and increase driver availability to reduce customer cancellations and enhance user experience. Additionally, understanding the specific reasons behind system rejections of orders is crucial for service optimization.

### 2

```python
# 创建一个新列来表示订单的小时
data_orders['order_hour'] = [data.split(':')[0] for data in data_orders['order_datetime']]

# 根据订单状态和小时数对订单进行分组，并计算每个状态的订单数量
order_counts = data_orders.groupby(['order_hour', 'order_status_key', 'is_driver_assigned_key']).size()

order_counts_df = order_counts.reset_index(name='counts')

plt.figure(figsize=(14, 7))
label = ["Client_canceled_before_driver_assigned", "System_canceled_before_driver_assigned",
         "Client_canceled_after_driver_assigned", "System_canceled_after_driver_assigned"]
# 为每个状态绘制一条线
i = 0
for status, df in order_counts_df.groupby(['order_status_key', 'is_driver_assigned_key']):
    plt.plot(df['order_hour'], df['counts'], "o-",
             label=label[i])
    i += 1

plt.title('Distribution of Failed Orders by Hours')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Failed Orders')
plt.legend(title='Order Status')

plt.grid(True)
plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域

plt.show()
```



1. **Sharp Increase in Order Failures During Morning Peak**: There is a notable surge in the number of failed orders around 8 AM. This could be associated with the morning rush hour when there is a significant rise in the demand for transportation services, possibly leading to an increased rate of cancellations and rejections due to longer response times or a shortage of available drivers.

2. **Dips in Order Failures During Late Night and Mid-Afternoon**: The number of failed orders experiences a decline during the late-night hours around 4 AM and the mid-afternoon around 2 PM. These periods are likely times of relatively lower demand for rides, resulting in a reduction in the total volume of orders and, consequently, fewer failed orders.

3. **Resurgence in Order Failures in the Evening**: From 5 PM onwards, there is a resurgence in the number of failed orders, with a minor peak around 7 PM. This likely corresponds to another peak period of demand as individuals are either heading home from work or preparing to go out for the evening.

4. **Stabilization of Order Failures in the Late Night**: After 10 PM, the number of failed orders tends to stabilize, with a slight increase towards midnight. This stability may indicate a balanced supply and demand for transportation services during these late-night hours. These observations suggest that strategies to reduce the rate of failed orders could be focused on peak demand periods, particularly during the morning rush hour. Potential improvements might include increasing the pool of available drivers during peak times or offering incentives for customers to schedule rides outside of these peak periods. Furthermore, gaining a deeper understanding of the specific reasons behind cancellations and rejections during these peak times is crucial for effectively addressing the root causes of these issues.

```python
data_orders['order_hour'] = pd.to_datetime(data_orders['order_datetime']).dt.hour
```

```python
cancelled_orders = data_orders[data_orders['order_status_key'] == 4]
plt.figure(figsize=(14, 7))

# Boxplot for cancellations with a driver
plt.subplot(1, 2, 1)
cancelled_orders[cancelled_orders['is_driver_assigned_key'] == 1]['cancellations_time_in_seconds'].plot(kind='box')
plt.title('Cancellation Time with Driver')

# Boxplot for cancellations without a driver
plt.subplot(1, 2, 2)
cancelled_orders[cancelled_orders['is_driver_assigned_key'] == 0]['cancellations_time_in_seconds'].plot(kind='box')
plt.title('Cancellation Time without Driver')

plt.tight_layout()
plt.show()
```


- Both charts reveal several outliers, which are the data points that lie beyond the whiskers of the boxplots, representing orders with unusually long cancellation durations.
- Compared to "Cancellation Time without Driver," the "Cancellation Time with Driver" exhibits a broader range of outliers. This could suggest that there is greater variability in the waiting period customers are willing to endure before canceling once a driver has been assigned.
- The central box in both plots, indicative of the interquartile range (IQR), is quite tight, implying that the majority of cancellations occur within a relatively concise and uniform timeframe.

```python
# 去除异常值，这里使用1.5倍的IQR作为异常值的判断标准
Q1 = data_orders['cancellations_time_in_seconds'].quantile(0.25)
Q3 = data_orders['cancellations_time_in_seconds'].quantile(0.75)
IQR = Q3 - Q1
outlier_filter = ((data_orders['cancellations_time_in_seconds'] < (Q1 - 1.5 * IQR)) | 
                  (data_orders['cancellations_time_in_seconds'] > (Q3 + 1.5 * IQR)))
data_orders = data_orders[~outlier_filter]
not_assigned_group = data_orders.groupby("is_driver_assigned_key").get_group(0)
```

```python

avg_time_without_driver = not_assigned_group.groupby("order_hour")['cancellations_time_in_seconds'].mean().reset_index()
do_assigned_group = data_orders.groupby("is_driver_assigned_key").get_group(1)
avg_time_with_driver = do_assigned_group.groupby("order_hour")['cancellations_time_in_seconds'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 8))
# 绘制平均值
plt.plot(avg_time_without_driver ['order_hour'], avg_time_without_driver ['cancellations_time_in_seconds'], marker='o', label="not assigned")
plt.plot(avg_time_with_driver['order_hour'],avg_time_with_driver['cancellations_time_in_seconds'], marker='o', label="assigned")

plt.title('Cancellation Time')
plt.legend()
plt.xlabel('Hour of the Day')
plt.ylabel('Average ETA (seconds)')
plt.grid(True)
plt.show()
```

The presence of outliers in cancellation timeframes may necessitate an in-depth exploration to discern the specific scenarios that result in extended waiting periods prior to cancellation.
The variability observed in cancellation timeframes when a driver is assigned suggests that customer patience may be influenced by factors such as the time of day and their unique circumstances.
The consistency in cancellation timeframes in the absence of a driver assignment could facilitate the prediction of customer behavior in situations where drivers are not immediately available.

```python
data_orderss = data_orders.dropna()
# 计算平均值
avg_time = data_orderss.groupby("order_hour")['m_order_eta'].mean().reset_index()

plt.figure(figsize=(10, 8))
# 绘制平均值
plt.plot(avg_time['order_hour'], avg_time['m_order_eta'], marker='o')
plt.grid(True)
plt.title('m_order_eta Time')
plt.xlabel('Hour of the Day')
plt.ylabel('ETA (seconds)')
plt.show()
```



1. **Fluctuations in Early Morning and Morning Peak ETA**: There is significant fluctuation in the average ETA between 4 AM and 7 AM, with an initial drop followed by an increase, likely due to changes in road conditions or a decrease in the number of available drivers. Additionally, at 8 AM, the ETA peaks, primarily as a result of traffic congestion during the morning rush hour.

2. **Changes in ETA from Morning to Afternoon**: By 10 AM, the average ETA decreases to a lower level, possibly due to improved traffic conditions and increased driver availability. Variability in ETA is observed from the afternoon into the evening, which may be related to fluctuations in traffic volume and uneven driver distribution throughout the day.

3. **Evening Rise in ETA**: There is another peak in ETA between 8 PM and 10 PM, followed by a decline starting at 11 PM. This evening peak may correspond to increased demand and traffic congestion as people return home from work or evening activities.

4. **Decline in Late Night ETA**: After 11 PM, the average ETA begins to drop, suggesting a reduction in traffic flow during the night, which in turn shortens travel times. This pattern indicates that Gett may need to adjust its vehicle dispatch strategy across different times of the day, particularly during traffic peaks, by increasing the number of drivers to reduce passenger wait times, and considering the impact of traffic congestion and other external factors to enhance service efficiency.

```python
import h3
import folium
import pandas as pd
import h3
import folium
from folium.plugins import HeatMap
from folium.features import DivIcon




# 设置 H3 的分辨率，对于大小为8的六边形，分辨率应该是8
h3_res = 8

# 使用 apply 函数为每个订单计算 H3 索引，并将其存储在新的 'h3_index' 列中

data_orders['h3_index'] = data_orders.apply(lambda x: h3.geo_to_h3(x['origin_latitude'], x['origin_longitude'], 8),axis=1)
# 点数儿，默认降序排列
# 计算每个 H3 索引的订单数量
order_counts = data_orders['h3_index'].value_counts()

# 确定包含80%订单的 H3 索引数量
total_orders = len(data_orders)
cumulative_orders = order_counts.cumsum()
indexes_80_percent =cumulative_orders[cumulative_orders <= total_orders * 0.8]




# 初始化一个地图
m = folium.Map(location=[data_orders['origin_latitude'].mean(), data_orders['origin_longitude'].mean()], zoom_start=12)

# 对于每个H3索引，生成一个多边形并将其添加到地图上
for h3_index in indexes_80_percent.index:
    # 获取H3索引的边界
    boundary = h3.h3_to_geo_boundary(h3_index, geo_json=True)
    # 创建一个多边形
    geojson = {
        "type": "Polygon",
        "coordinates": [boundary]
    }
    # 获取订单失败次数作为颜色深浅的依据
    fails_count = data_orders[data_orders['h3_index'] == h3_index]['order_status_key'].count()
    # 添加多边形到地图上，使用失败次数来决定颜色
    folium.GeoJson(geojson, style_function=lambda x, fails_count=fails_count: {
        'fillColor': '#ff0000' if fails_count > 100 else '#00ff00',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.6,
    }).add_to(m)

# 保存地图为一个HTML文件
m.save('hex_map.html')
```
