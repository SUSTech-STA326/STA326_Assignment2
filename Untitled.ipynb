```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orders = pd.read_csv('datasets/data_orders.csv')
offers = pd.read_csv('datasets/data_offers.csv')
```

```python
orders.head(5)
```

```python
offers.head(5)
```

```python
unique_values = orders['order_status_key'].value_counts()
print(unique_values)
```

# (1)

```python
merged_df = pd.merge(orders, offers, on='order_gk', how='left')
merged_df.head(5)
```

```python
conditions = [
    (merged_df['order_status_key'] == 4) & (merged_df['is_driver_assigned_key'] == 1),
    (merged_df['order_status_key'] == 4) & (merged_df['is_driver_assigned_key'] == 0),
    (merged_df['order_status_key'] == 9) & (merged_df['is_driver_assigned_key'] == 1),
    (merged_df['order_status_key'] == 9) & (merged_df['is_driver_assigned_key'] == 0)
]

# Labels for the conditions
labels = [
    'Cancelled by client, Driver assigned',
    'Cancelled by client, No driver',
    'Cancelled by system, Driver assigned',
    'Cancelled by system, No driver'
]

# Apply conditions to create a new column for cancellation categories
merged_df['Cancellation_Category'] = np.select(conditions, labels, default='Other')
merged_df.head(5)
```

```python
# Count the occurrences of each cancellation category
cancellation_counts = merged_df['Cancellation_Category'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
cancellation_counts.plot(kind='bar')
plt.title('Cancellation Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

cancellation_counts
```

# (2)

```python
# to_datetime
# merged_df['order_datetime'] = pd.to_datetime(merged_df['order_datetime'], format='%H:%M:%S')
# merged_df['order_hour'] = merged_df['order_datetime'].dt.hour

# strapplyint
merged_df['order_hour'] = merged_df['order_datetime'].str[0:2].apply(lambda x: int(x))
merged_df['order_min'] = merged_df['order_datetime'].str[3:5].apply(lambda x: int(x))
merged_df['order_second'] = merged_df['order_datetime'].str[6:8].apply(lambda x: int(x))
merged_df.head(5)
```

## 

```python
# 
hourly_cancellation_counts = merged_df.groupby(['order_hour', 'Cancellation_Category']).size().unstack(fill_value=0)

# 
plt.figure(figsize=(14, 8))
hourly_cancellation_counts.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Distribution of Failed Orders by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Failed Orders')
plt.xticks(rotation=0)
plt.legend(title='Cancellation Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 

```python
merged_df['seconds_since_midnight'] = merged_df['order_hour']*3600 + merged_df['order_min']*60 + merged_df['order_second']
merged_df.head(5)
```

```python
merged_df['30min_interval'] = merged_df['seconds_since_midnight'] // (30 * 60) + 1
merged_df.head(5)
```

```python
# 30
interval_cancellation_counts = merged_df.groupby(['30min_interval', 'Cancellation_Category']).size().unstack(fill_value=0)
interval_cancellation_counts.head(5)
# interval_cancellation_counts.index
# interval_cancellation_counts.columns
```

```python
# X
time_labels = [f"{int(i/2)}:{('00' if i % 2 == 0 else '30')}" for i in range(48)]

# 
plt.figure(figsize=(14, 8))
for category in interval_cancellation_counts.columns:
    plt.plot(interval_cancellation_counts.index, interval_cancellation_counts[category], label=category)

plt.title('Failed Orders Over Time of Day (30min Intervals)')
plt.xlabel('Time of Day')
plt.ylabel('Number of Failed Orders')
plt.xticks(interval_cancellation_counts.index, time_labels, rotation=45)
# plt.xticks(interval_cancellation_counts.index, rotation=45)
plt.legend(title='Cancellation Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  - IQR
Q1 = merged_df['cancellations_time_in_seconds'].quantile(0.25)
Q3 = merged_df['cancellations_time_in_seconds'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 
filtered_merged_df = merged_df[(merged_df['cancellations_time_in_seconds'] >= lower_bound) & 
                               (merged_df['cancellations_time_in_seconds'] <= upper_bound)]

print(len(merged_df), len(filtered_merged_df))
```

```python
# 
grouped_avg_cancellation_time = filtered_merged_df.groupby(['is_driver_assigned_key', 'order_hour'])['cancellations_time_in_seconds'].mean().unstack(0)
grouped_avg_cancellation_time
```

```python
# 
plt.figure(figsize=(14, 8))
grouped_avg_cancellation_time.plot(marker='o')
plt.title('Average Cancellation Time by Hour (Filtered)')
plt.ylabel('Average Cancellation Time (seconds)')
plt.xlabel('Hour of Day')
plt.xticks(range(0, 24))
plt.legend(['Without Driver', 'With Driver'], title='Driver Assigned')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import h3
import folium
import branca.colormap as cm
```

```python
# orders_dfDataFrame
# h3
orders['h3_code'] = orders.apply(lambda x: h3.geo_to_h3(x['origin_latitude'], x['origin_longitude'], 8), axis=1)
orders.head(5)
```

```python
# 
h3_counts = orders['h3_code'].value_counts()
h3_counts
```

```python
# 80%
total_orders = h3_counts.sum()
cumulative_counts = h3_counts.cumsum()
threshold = total_orders * 0.8
relevant_hexes = cumulative_counts[cumulative_counts <= threshold]
relevant_hexes
```

```python
map_center = [orders['origin_latitude'].mean(), orders['origin_longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=14)
# foliummzoom_start
```

## 

```python
# relevant_hexes
for hex_id in relevant_hexes.index:
    # 
    boundary = h3.h3_to_geo_boundary(hex_id, geo_json=True)
    # folium
    boundary_coordinates = [[coord[::-1] for coord in boundary]] # 
    # 
    folium.Polygon(boundary_coordinates,
                   color='blue',
                   weight=1,
                   fill=True,
                   fill_color='blue',
                   fill_opacity=0.5,
                   tooltip=f'Hex ID: {hex_id}').add_to(m)
    
    # print(boundary, boundary_coordinates)

m.save('hex_map.html')
```

## count

```python
# 
max_count = h3_counts.max()
min_count = h3_counts.min()

# branca
color_scale = cm.linear.YlOrRd_09.scale(min_count, max_count)

# 
color_dict = {hex_id: color_scale(count) for hex_id, count in h3_counts.items()}

# relevant_hexes
for hex_id in relevant_hexes.index:
    # 
    boundary = h3.h3_to_geo_boundary(hex_id, geo_json=True)
    # folium
    boundary_coordinates = [[coord[::-1] for coord in boundary]]  # 
    # 
    hex_color = color_dict[hex_id]
    # 
    folium.Polygon(boundary_coordinates,
                   color=hex_color,
                   weight=1,
                   fill=True,
                   fill_color=hex_color,
                   fill_opacity=0.7,
                   tooltip=f'Hex ID: {hex_id}, Count: {h3_counts[hex_id]}').add_to(m)

m.save('hex_map_colored.html')
```

```python
# relevant_hexes
for hex_id in relevant_hexes.index:
    polygons = h3.h3_to_geo_boundary(hex_id, geo_json=True)
    # 
    folium.Polygon(locations=polygons, color='green', fill=True, fill_color='green', fill_opacity=1).add_to(m)

m.save('hex_map.html')
```
