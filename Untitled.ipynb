```python
import matplotlib.pyplot as plt

# 读取数据文件到DataFrame
df = orders

# 使用groupby()函数按照'order_status_key'和'is_driver_assigned_key'进行分组，并计算频率
grouped = df.groupby(['order_status_key', 'is_driver_assigned_key']).size().reset_index(name='count')

# 绘制柱状图
plt.figure(figsize=(8, 6))

# 遍历每种可能的组合
for idx, row in grouped.iterrows():
    plt.bar(f"{row['order_status_key']} - {row['is_driver_assigned_key']}", row['count'], label=f"{row['order_status_key']} - {row['is_driver_assigned_key']}")

plt.xlabel('combination')
plt.ylabel('count')
plt.title('Q1')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4 - cancelled by client
# 9 - cancelled by system
# 1 the driver has been assigned
# 0 the driver hasn't been assigned
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 绘制箱线图
plt.figure(figsize=(8, 6))
sns.boxplot(x=orders["cancellations_time_in_seconds"])
plt.title('Order Cancellation Time Boxplot')
plt.xlabel('Cancellation Time (seconds)')
plt.show()
```

```python
# 计算异常值
Q1 = orders['cancellations_time_in_seconds'].quantile(0.25)
Q3 = orders['cancellations_time_in_seconds'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 打印异常值
outliers = orders[(orders['cancellations_time_in_seconds'] < lower_bound) | (orders['cancellations_time_in_seconds'] > upper_bound)]

# 画出异常值分布的直方图
plt.hist(outliers['cancellations_time_in_seconds'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Outliers')
plt.xlabel('Cancellation Time (Seconds)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设Q3是一个DataFrame，包含了你的数据
Q3 = orders[orders['cancellations_time_in_seconds'] <= upper_bound]
Q3 = orders[orders['order_status_key'] == 4]
# 将order_datetime列转换为datetime类型
Q3.loc[:, "order_datetime"] = pd.to_datetime(Q3["order_datetime"])

# 筛选出B列为1和0的数据
B1_data = Q3[Q3['is_driver_assigned_key'] == 1]
B0_data = Q3[Q3['is_driver_assigned_key'] == 0]

# 计算每小时的平均钱的数量
B1_hourly_mean = B1_data.groupby(B1_data['order_datetime'].dt.hour)['cancellations_time_in_seconds'].mean()
B0_hourly_mean = B0_data.groupby(B0_data['order_datetime'].dt.hour)['cancellations_time_in_seconds'].mean()

# 创建图像
plt.figure(figsize=(10, 6))

# 画出两条曲线
plt.plot(B1_hourly_mean.index, B1_hourly_mean.values, marker='o', linestyle='-', label='is_driver_assigned_key=1')
plt.plot(B0_hourly_mean.index, B0_hourly_mean.values, marker='o', linestyle='-', label='is_driver_assigned_key=0')

# 添加标题和标签
plt.title('Hourly Average time Quantity')
plt.xlabel('Hour of Day')
plt.ylabel('Average time Quantity')

# 添加图例
plt.legend()

# 显示图像
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import folium
import h3
from matplotlib.colors import LinearSegmentedColormap, rgb2hex
from matplotlib.colors import LinearSegmentedColormap

# 使用h3库将经纬度转换为h3索引，并添加到数据表中
def convert_to_h3(row):
    return h3.geo_to_h3(row['origin_latitude'], row['origin_longitude'], 8)

orders['h3_index'] = orders.apply(convert_to_h3, axis=1)

# 对订单数据按照h3索引进行分组，计算每个h3索引内订单的数量
h3_counts = orders.groupby('h3_index').size().reset_index(name='order_count')

# 按订单数量降序排列
h3_counts = h3_counts.sort_values(by='order_count', ascending=False)

# 计算80%的订单数量
total_orders = h3_counts['order_count'].sum()
threshold_order_count = 0.8 * total_orders

# 找到包含80%订单的h3索引
cumulative_order_count = 0
selected_h3_indexes = []
for index, row in h3_counts.iterrows():
    cumulative_order_count += row['order_count']
    if cumulative_order_count <= threshold_order_count:
        selected_h3_indexes.append(row['h3_index'])
    else:
        break

# 创建颜色映射
cmap = LinearSegmentedColormap.from_list(name='custom_cmap', colors=['red',  'yellow', 'blue'])

# 创建地图对象
m = folium.Map(location=[orders['origin_latitude'].mean(), orders['origin_longitude'].mean()], zoom_start=10)

# 在地图上画出选定的h3索引，并根据订单数量进行着色
for h3_index in selected_h3_indexes:
    boundary = h3.h3_to_geo_boundary(h3_index)
     # 将边界坐标的第一个点添加到列表末尾，以闭合多边形
    boundary = list(boundary)  # 转换为列表以便修改
    boundary.append(boundary[0])
    order_count = h3_counts.loc[h3_counts['h3_index'] == h3_index, 'order_count'].iloc[0]
    color = cmap(order_count / threshold_order_count)
    folium.Polygon(locations=boundary, fill=True, color=rgb2hex(color), fill_opacity=0.7).add_to(m)

# 显示地图
display(m)
```
